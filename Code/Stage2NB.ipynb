```julia
Data = CSV.read(datapath * "Data_1.csv", DataFrame);
SelectedProds = load(SelectedDataPath * "SelectedProds.jld")["SelectedProds"];
SelectedData = [CSV.read(SelectedDataPath * "$i.csv", DataFrame) for i ∈ SelectedProds];
```

```julia
PricesDataFrame = CSV.read(ExportedData1Path * "PricesMatrix.csv", DataFrame);
PricesMatrix = load(ExportedData1Path * "PricesMatrix.jld")["PricesMatrix"];
```

```julia
GoldPrices = PricesDataFrame."1-20" #1-20 is category of gold
NormalizedPrices = DataFrame()
for i ∈ 1:length(SelectedProds)
    NormalizedPrices[!, SelectedProds[i]] = PricesDataFrame[!, SelectedProds[i]] ./ GoldPrices
end
```

```julia
NormalizedPricesMatrix = Array(NormalizedPrices);
CorMatPrices = cor(NormalizedPricesMatrix);
CovMatPrices = cov(NormalizedPricesMatrix);
```

```julia
Names = CostumMods.GetProdName.(parse.(Int, getindex.(split.(SelectedProds, "-"), 2)))
display(heatmap(nothing, Names, CorMatPrices, xticks=1:17, c=cgrad(rev=true), size = (1000, 600)))
# display(heatmap(CovMatPrices, xticks=1:17, yticks=1:17, ratio=:equal, c=cgrad(rev=true)))
```
